```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Simple Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize=(8,5), dpi=100)

plt.plot(x,y, 'b^--', label='2x')

x2 = np.arange(0,4.5,0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,])

plt.legend()

plt.savefig('mygraph.png', dpi=300)
plt.show()
```

## Bar Chart

```python

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

# data Visulations on Laptop Price Predications

```python
# Group laptops by 'processor_name' and compute average price
df= pd.read_csv('laptopPrice.csv')
cpu_price_comparison = df.groupby('processor_name')['Price'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
cpu_price_comparison.plot(kind='bar', color='skyblue')
plt.title('Average Laptop Price by Processor Name')
plt.ylabel('Average Price (INR)')
plt.xlabel('Processor Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Histogram

```python
price_bins = [20000, 30000, 40000, 50000, 60000, 70000, 80000, 100000]

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(df['Price'], bins=price_bins, color='#abcdef', edgecolor='black')
plt.xticks(price_bins, rotation=45)
plt.ylabel('Number of Laptops')
plt.xlabel('Price (INR)')
plt.title('Distribution of Laptop Prices')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('laptop_price_histogram.png', dpi=300)
plt.show()
```

### Pie Chart 1

```python
touch_yes = df[df['Touchscreen'] == 'Yes'].count()[0]
touch_no = df[df['Touchscreen'] == 'No'].count()[0]

# Pie chart
plt.figure(figsize=(8, 5))
labels = ['Touchscreen', 'Non-Touchscreen']
colors = ['#abcdef', '#aabbcc']
plt.pie([touch_yes, touch_no], labels=labels, colors=colors, autopct='%.2f %%')
plt.title('Touchscreen vs Non-Touchscreen Laptops')
plt.show()
```

### Pie Chart 2

```python
cheap = df[df['Price'] < 30000].count()[0]
mid_low = df[(df['Price'] >= 30000) & (df['Price'] < 40000)].count()[0]
mid = df[(df['Price'] >= 40000) & (df['Price'] < 50000)].count()[0]
mid_high = df[(df['Price'] >= 50000) & (df['Price'] < 70000)].count()[0]
premium = df[df['Price'] >= 70000].count()[0]

# Data for pie chart
prices = [cheap, mid_low, mid, mid_high, premium]
labels = ['Under 30K', '30K-40K', '40K-50K', '50K-70K', 'Above 70K']
explode = (0.3, 0.1, 0, 0, 0.3)

# Plot
plt.figure(figsize=(8, 5), dpi=100)
plt.style.use('ggplot')
plt.title('Laptop Price Range Distribution (INR)')
plt.pie(prices, labels=labels, explode=explode, pctdistance=0.8, autopct='%.2f %%')
```

### Box and Whiskers Chart

```python
top_brands = df['brand'].value_counts().nlargest(3).index.tolist()

# Create price groups for each brand
brand1 = df[df['brand'] == top_brands[0]]['Price']
brand2 = df[df['brand'] == top_brands[1]]['Price']
brand3 = df[df['brand'] == top_brands[2]]['Price']

# Boxplot
plt.figure(figsize=(5, 8), dpi=100)
plt.style.use('default')
bp = plt.boxplot(
    [brand1, brand2, brand3],
    labels=top_brands,
    patch_artist=True,
    medianprops={'linewidth': 2}
)

plt.title('Laptop Price Comparison by Brand')
plt.ylabel('Price (INR)')

# Customizing box appearance
for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor='#e0e0e0')

plt.show()
```
